```python
########### REGRESIN MULTIVARIANTE PARA EFECTO ENERO ###########
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
import numpy as np
from datetime import datetime
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns

# Fechas de inicio y fin
start = datetime(1990, 1, 1)
end = datetime.now()

# Acciones a analizar
stocks = ['AAPL', 'AMZN', 'TSLA', 'NVDA', 'META', 'GOOGL', 'MSFT', 'SPY']

# Obtiene los datos de las acciones
try:
    data = yf.download(stocks, start=start, end=end)
except Exception as e:
    print(f"Error al descargar datos: {e}")

# Calcular los retornos mensuales
data_monthly = data['Adj Close'].resample('M').ffill()
returns = np.log(data_monthly) - np.log(data_monthly.shift(1))
returns = returns.dropna()

# Crear una variable dummie para enero
returns['Enero'] = (returns.index.month == 1).astype(int)

# Crear un DataFrame para almacenar los resultados extendidos
extended_summary_data = {
    'Accin': [],
    'Intercepto': [],
    'Coeficiente_Enero': [],
    'Coeficiente_SPY': [],
    'P_valor_Enero': [],
    'P_valor_SPY': [],
    'Error_cuadrtico_medio': [],
    'Error_absoluto_medio': [],
    'R_cuadrado': [],
    'SSR': [],
    'Desviacin_Residuos': []
}

# Realiza la regresin para cada accin y calcula las estadsticas adicionales
for stock in stocks[:-1]:  # Excluir SPY de la lista de acciones a analizar
    y = returns[stock]
    X = sm.add_constant(returns[['Enero', 'SPY']])  # Aade SPY como variable independiente
    
    model = sm.OLS(y, X)
    results = model.fit()
    
    # Calcular el MSE y MAE usando validacin cruzada
    kf = KFold(n_splits=5, random_state=42, shuffle=True)
    mse_values = []
    mae_values = []
    ssr_values = []
    std_resid_values = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        model_cv = sm.OLS(y_train, X_train)
        results_cv = model_cv.fit()
        
        y_pred_test = results_cv.predict(X_test)
        
        mse = mean_squared_error(y_test, y_pred_test)
        mae = mean_absolute_error(y_test, y_pred_test)
        ssr = np.sum((y_test - y_pred_test) ** 2)
        std_resid = np.std(y_test - y_pred_test)
        
        mse_values.append(mse)
        mae_values.append(mae)
        ssr_values.append(ssr)
        std_resid_values.append(std_resid)
    
    extended_summary_data['Accin'].append(stock)
    extended_summary_data['Intercepto'].append(results.params[0])
    extended_summary_data['Coeficiente_Enero'].append(results.params['Enero'])
    extended_summary_data['Coeficiente_SPY'].append(results.params['SPY'])
    extended_summary_data['P_valor_Enero'].append(results.pvalues['Enero'])
    extended_summary_data['P_valor_SPY'].append(results.pvalues['SPY'])
    extended_summary_data['Error_cuadrtico_medio'].append(np.mean(mse_values))
    extended_summary_data['Error_absoluto_medio'].append(np.mean(mae_values))
    extended_summary_data['R_cuadrado'].append(results.rsquared)
    extended_summary_data['SSR'].append(np.sum((results.resid) ** 2))
    extended_summary_data['Desviacin_Residuos'].append(np.std(results.resid))

# Crear el DataFrame de resumen extendido
extended_summary_df = pd.DataFrame(extended_summary_data)

# Imprimir el resumen completo del modelo con las estadsticas adicionales
print("Resumen extendido del modelo:")
print(extended_summary_df[['Accin', 'Intercepto', 'Coeficiente_Enero', 'Coeficiente_SPY', 'P_valor_Enero', 'P_valor_SPY', 'Error_cuadrtico_medio', 'Error_absoluto_medio', 'R_cuadrado', 'SSR', 'Desviacin_Residuos']])

# Anlisis exploratorio
correlation_matrix = returns.corr()
print("Matriz de correlacin:")
print(correlation_matrix)

# Graficar correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlacin de Retornos')
plt.show()

print(stocks)
print(returns)
```

```python
########### REGRESIN MULTIVARIANTE RALLY NAVIDAD ###########
import pandas as pd
import yfinance as yf
import numpy as np
from datetime import datetime
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt

# Fechas de inicio y fin
start = datetime(1990, 1, 1)
end = datetime.now()

# Acciones a analizar
stocks = ['AAPL', 'AMZN', 'TSLA', 'NVDA', 'META', 'GOOGL', 'MSFT', 'SPY']

# Obtiene los datos de las acciones
data = yf.download(stocks, start=start, end=end)['Adj Close']

# Calcular los rendimientos mensuales
returns = np.log(data.resample('M').ffill()) - np.log(data.resample('M').ffill().shift(1))
returns.dropna(inplace=True)

# Crear una variable dummy para noviembre y diciembre
returns['Nov_Dic'] = ((returns.index.month == 11) | (returns.index.month == 12)).astype(int)

# Aadir SPY a las variables independientes con desplazamiento
returns['SPY'] = returns['SPY'].shift(1)
returns.dropna(inplace=True)

# Crear una matriz de correlaciones
correlation_matrix = returns.corr()

# Visualizar la matriz de correlaciones como un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones')
plt.show()

# Crear un DataFrame para almacenar los resultados extendidos
extended_summary_data = {
    'Accin': [],
    'Coeficiente_Nov_Dic': [],
    'Coeficiente_SPY': [],
    'P_valor_Nov_Dic': [],
    'P_valor_SPY': [],
    'Error_cuadrtico_medio': [],
    'Error_absoluto_medio': [],
    'R_cuadrado': [],
    'SSR': [],
    'Desviacin_Residuos': []
}

# Iterar sobre cada accin para ajustar el modelo y obtener los resultados
for stock in stocks[:-1]:  # Excluir SPY de la lista de acciones a analizar
    X = sm.add_constant(returns[['Nov_Dic', 'SPY']])  # Aadir Nov_Dic y SPY como variables independientes
    y = returns[stock]
    
    model = sm.OLS(y, X)
    results = model.fit()
    
    coef_Nov_Dic = results.params['Nov_Dic']
    coef_SPY = results.params['SPY']
    p_value_Nov_Dic = results.pvalues['Nov_Dic']
    p_value_SPY = results.pvalues['SPY']
    
    y_pred = results.predict(X)
    
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    
    extended_summary_data['Accin'].append(stock)
    extended_summary_data['Coeficiente_Nov_Dic'].append(coef_Nov_Dic)
    extended_summary_data['Coeficiente_SPY'].append(coef_SPY)
    extended_summary_data['P_valor_Nov_Dic'].append(p_value_Nov_Dic)
    extended_summary_data['P_valor_SPY'].append(p_value_SPY)
    extended_summary_data['Error_cuadrtico_medio'].append(mse)
    extended_summary_data['Error_absoluto_medio'].append(mae)
    extended_summary_data['R_cuadrado'].append(results.rsquared)
    extended_summary_data['SSR'].append(np.sum(results.resid ** 2))
    extended_summary_data['Desviacin_Residuos'].append(np.std(results.resid))

# Crear el DataFrame de resumen extendido
extended_summary_df = pd.DataFrame(extended_summary_data)

# Imprimir el resumen completo del modelo con las estadsticas adicionales
print("\nResumen extendido del modelo:")
print(extended_summary_df[['Accin', 'Coeficiente_Nov_Dic', 'Coeficiente_SPY', 'P_valor_Nov_Dic', 'P_valor_SPY', 'Error_cuadrtico_medio', 'Error_absoluto_medio', 'R_cuadrado', 'SSR', 'Desviacin_Residuos']])
```

```python
######## RETORNOS MEDIOS DIARIOS, VOLATILIDAD DIARIA Y JARQUE-BERA #######
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis, jarque_bera

# Lista de tickers
tickers = ['AAPL', 'AMZN', 'TSLA', 'MSFT', 'META', 'GOOGL', 'NVDA',"SPY"]

start_date = '1990-01-01'
end_date = '2024-12-31'

# Obtener los das de la semana en orden cronolgico
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Crear un DataFrame vaco para almacenar los retornos diarios promedio de todos los tickers
average_returns = pd.DataFrame(index=ordered_days)

# Crear un DataFrame para almacenar las medidas estadsticas
stats_data = {'Ticker': [], 'Day': [], 'Mean': [], 'Volatility': [], 'Jarque-Bera': []}

for ticker in tickers:
    # Descargar datos histricos
    data = yf.download(ticker, start=start_date, end=end_date)

    # Calcular los retornos diarios logartmicos
    data['Daily_Return'] = np.log(data['Adj Close'] / data['Adj Close'].shift(1))

    # Eliminar el primer registro NaN resultante de los retornos diarios
    data = data.dropna()

    # Calcular el retorno diario promedio y aadirlo al DataFrame
    data['Day'] = pd.Categorical(data.index.strftime('%A'), categories=ordered_days, ordered=True)
    average_returns[ticker] = data.groupby('Day')['Daily_Return'].mean()

    # Calcular las medidas estadsticas para cada da de la semana
    for day in ordered_days:
        daily_returns = data[data['Day'] == day]['Daily_Return']
        stats_data['Ticker'].append(ticker)
        stats_data['Day'].append(day)
        stats_data['Mean'].append(daily_returns.mean())
        stats_data['Volatility'].append(daily_returns.std())
        stats_data['Jarque-Bera'].append(jarque_bera(daily_returns)[0])

# Crear un DataFrame para las medidas estadsticas
stats_df = pd.DataFrame(stats_data)

# Crear un grfico de barras para visualizar los retornos diarios promedio
average_returns.plot(kind='bar', figsize=(12, 6))

plt.title('Historical return')
plt.xlabel('Day of the Week')
plt.ylabel('Average return')
plt.legend(title='Ticker')
plt.show()

# Imprimir el resumen estadstico de los retornos diarios
print("Resumen estadstico de los retornos diarios:")
print(stats_df)

# Crear un grfico de barras para la volatilidad
plt.figure(figsize=(12, 6))
sns.barplot(x='Day', y='Volatility', hue='Ticker', data=stats_df)
plt.title('Volatility by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Volatility')

# Aadir una lnea constante para la volatilidad como la media de todas las acciones por da
average_volatility = stats_df['Volatility'].mean()
plt.axhline(y=average_volatility, color='r', linestyle='--')
plt.show()
```

```python
####### REGRESIN MULTIVARIANTE EFECTO FIN DE SEMANA #######
import pandas as pd
import yfinance as yf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Fechas de inicio y fin
start = datetime(1990, 1, 1)
end = datetime.now()

# Acciones a analizar
stocks = ['AAPL', 'AMZN', 'TSLA', 'NVDA', 'META', 'GOOGL', 'MSFT', 'SPY']

# Obtiene los datos de las acciones
data = yf.download(stocks, start=start, end=end)['Adj Close']

# Calcular los rendimientos mensuales
returns = np.log(data.resample('M').ffill()) - np.log(data.resample('M').ffill().shift(1))
returns.dropna(inplace=True)

# Fusionar viernes y lunes en un mismo coeficiente como dummy
returns['Viernes_Lunes'] = ((returns.index.dayofweek == 4) | (returns.index.dayofweek == 0)).astype(int)  # 4 es Viernes, 0 es Lunes

# Correlaciones de los retornos
correlation_matrix = returns.corr()

# Mapa de calor de correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlaciones de Retornos')
plt.show()

# Crear un DataFrame para almacenar los resultados
summary_data = {
    'Accin': [],
    'Intercepto': [],
    'Coeficiente_Viernes_Lunes': [],
    'Coeficiente_SPY': [],
    'P_valor_Viernes_Lunes': [],
    'P_valor_SPY': [],
    'Error_cuadrtico_medio': [],
    'Error_absoluto_medio': [],
    'R_cuadrado': [],
    'SSR': [],
    'Desviacin_Residuos': []
}

# Iterar sobre cada accin para ajustar el modelo y obtener los resultados
for stock in stocks:
    X = sm.add_constant(returns[['Viernes_Lunes', 'SPY']])  # Aadir una constante (intercepto) al modelo
    y = returns[stock]
    
    model = sm.OLS(y, X)
    results = model.fit()
    
    y_pred = results.predict(X)
    
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    r_squared = results.rsquared
    ssr = np.sum(results.resid ** 2)
    std_resid = np.std(results.resid)
    
    summary_data['Accin'].append(stock)
    summary_data['Intercepto'].append(results.params['const'])
    summary_data['Coeficiente_Viernes_Lunes'].append(results.params['Viernes_Lunes'])
    summary_data['Coeficiente_SPY'].append(results.params['SPY'])
    summary_data['P_valor_Viernes_Lunes'].append(results.pvalues['Viernes_Lunes'])
    summary_data['P_valor_SPY'].append(results.pvalues['SPY'])
    summary_data['Error_cuadrtico_medio'].append(mse)
    summary_data['Error_absoluto_medio'].append(mae)
    summary_data['R_cuadrado'].append(r_squared)
    summary_data['SSR'].append(ssr)
    summary_data['Desviacin_Residuos'].append(std_resid)

# Crear el DataFrame de resumen
summary_df = pd.DataFrame(summary_data)

# Imprimir el resumen completo del modelo
print("Resumen del modelo:")
print(summary_df)
```
